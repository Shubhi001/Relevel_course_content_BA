### Q3 Classes, Method and Objects

Create a class "Student" with at  3 attributes - Name, Age, Marks
Also create a method "Result" which will say that " {Name} is passed" if marks are >40% and failed otherwise.

Create 3 students objects and run "Result" for all 3. 

```python
class student:
    def __init__(self, name,age,marks):
        self.name=name
        self.age=age
        self.marks=marks
        
    def result(self):
        if(self.marks>40):
            print(f"{self.name} is passed")
        else:
            print(f"{self.name} is failed")
            
student1 = student('Aryan',21,39)
student2 = student('Chitransh',24,49)
student3 = student('Happy',19,99)

student1.result()
student2.result()
student3.result()
```

### Q7 Polymorphism

Define a class bird with attributte "name" and  a method "flight" that prints " Almost all bird can fly ".

Define 3 child classes Sparrow, Parrot and Kiwi - all havind a method "flight" that says whether the bird can fly or not. 
( Kiwis cant fly).

Make an object each of all the 4 classes and run "flight" method. 

```python
class bird:
    Name = 'general bird'
    def flight(self):
        print('Almost all birds can fly')
        
class Sparrow:
    name = "Sparrow"
    def flight(self):
        print('f{self.name} can fly')

class Parrot:
    name = "Parrot"
    def flight(self):
        print('f{self.name} can fly')
        
class Kiwi:
    name = "Sparrow"
    def flight(self):
        print('f{self.name} can NOT fly')
        
bird_object = bird()      
Sparrow_object = Sparrow()
Parrot_object = Parrot()
Kiwi_object = Kiwi()

bird_object.flight()
Sparrow_object.flight()
Parrot_object.flight()
Kiwi_object.flight()
```

### Q8 Define a class "Operations" which can take an input string and perform following functions - 
1. Print the reversed string word by word
2. Print the string in all UPPER case
3. Print the string in all LOWER case

( All these 3 functions will be performed by 3 methods defined within class Operations.) 

Perform all these functions on the string - " The quick brown fox jumps over the lazy dog "

```python
class Operations:
    
    def __init__(self,input_string):
        self.input=input_string
        
    def reverse_fn(self):
        split_input = self.input.split()
        reversed_input = reversed(split_input)
        print(' '.join(reversed_input))
            
    def upper_fn(self):
        print(self.input.upper())
            
    def lower_fn(self):
        print(self.input.lower())

        
String1 = Operations('The quick brown fox jumps over the lazy dog ')
String1.reverse_fn()
String1.upper_fn()
String1.lower_fn()
```

### Q9 Define a class "Cuboid_functions" which can take length, breadth and height as  input parameters and perform following functions - 

1. Calculate Surface area of cuboid
2. Calculate Volume of cuboid
3. Check if the shape is a Cube
( All these 3 functions will be performed by 3 methods defined within class Cuboid_functions.)

Perform all these functions on the input - "20,15,10"

```python
class Cuboid_functions:
    
    def __init__(self,l,b,h):
        self.l=l
        self.b=b
        self.h=h
        
    def surface_area(self):
        area = 2*((self.l*self.b)+(self.b*self.h)+(self.h*self.l))
        print(f'Surface are is {area}')
            
    def volume(self):
        volume = self.l*self.b*self.h
        print(f'Volume is {volume}')
            
    def is_cube(self):
        if(self.l==self.b & self.b==self.h):
            print('Shape is a Cube')
        else:
            print('Shape is NOT a cube')

Cube1 = Cuboid_functions(20,15,10)
Cube2 = Cuboid_functions(10,10,10)

Cube1.surface_area()
Cube1.volume()
Cube1.is_cube()

Cube2.surface_area()
Cube2.volume()
Cube2.is_cube()
```

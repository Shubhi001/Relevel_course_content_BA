### Advanced Panda tools - pandas.plotting

### Scatter Matrix

A scatter matrix visualises each variable in the dataset with another, in a pairwise fashion - 

```python
pd.plotting.scatter_matrix(iris);
plt.tight_layout()
```

Pair plot is used to understand 
the best set of features to explain a relationship between two variables or to form the most separated clusters. 

This is extremely useful in feature selection or rejection because we can see all the patterns among data 
instantly. It is also used to choose best set of variables for a classification problem. 

### Parallel Coordinates

Parallel coordinates are another great way to analyse multivariate data and find patterns.

Here, each of the variable is represented as parallel lines and data points are plotted and connected. 
The color coding helps us visualise the data points with respect to a target categorical variable. 


```python
plt.figure()
pd.plotting.parallel_coordinates(iris,'species');
```

Parallel coordinates are similar to time-series plots except that there is no time on the x-axis and there is no 
linear relationship among points on x-axis. 

These have applications to visualise very high dimensional data.

### Distplot

Distplot is used for checking the distribution of a continuous variable. 
It is a histogram, with a KDE plot layered on top pf hitogram. 

Dsiplot function of Seaborn library can be used to build this plot - 

```python
sns.distplot(iris['sepal_width'], hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'},bins=20);
```

### Jointplot

Jointplot combines Univariate and Bivariate analysis in a unique way. 

The relationship between 2 variables is shown in the plot, with histograms of individual variable on the axes. 

The "Kind" parameter helps in specifying one of the many alterations of this plot, with "Hex" being a commonly used one. 
Using "Hex" bins the data into hexagonal shapes on the chart as shown in example below. 

```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'], alpha=0.4);
```

```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'], alpha=0.4, kind ='hex');
```

'Reg' in the 'kind' parameter plots a regression line. 
We had also talked about how Seaborn has support for Linear Regression, this is one example for the same. 

```python
sns.jointplot(iris['sepal_length'],iris['sepal_width'], kind ='reg');
```

'kde' in the 'kind' parameter plots KDE in 2D and 1D on the plot and axes. 

```python
# set the seaborn style for all the following plots
sns.set_style('white')
sns.jointplot(iris['sepal_length'],iris['sepal_width'], alpha=0.4,kind='kde', space=0);
```

### Pairplot

Pairplots are analogous in Seaborn to Scatter Matrix discussed earlier. 

It plots each variable against another in a pairwise manner. 

A 'hue' parameter allows for categorising the data-point based on a target variable for a much more insightful chart. 

'diag_kind' dictates what kind of plot should be shown on diagonals where it's a univariate chart. 

```python
sns.pairplot(iris, hue='species', diag_kind='kde', size=2);
```

### Swarmplot and Violinplot

Swarmplot and Violinplot are similar to Boxplots, except that they provide addtional information. 

A Swarmplot is a box plot, but all the data points are also shown as dots in a symmetrical fashion. 

A Violinplot is a box plot, but it alos plots KDE graph of the variable. 

```python
plt.figure(figsize=(8,6));
plt.subplot(121)
sns.swarmplot('species', 'petal_length', data=iris);
plt.subplot(122)
sns.violinplot('species', 'petal_length', data=iris);
```

Spurious Correlation is a situation when 2 variables appear to be extremely related, but in reality they are not.

### Editing Charts for Seaborn plots

We know that Seaborn is nothing but an another scripting layer on top of Matplotlib and Pyplot.
Hence we can use ether use Axis object as well as pyplot functions discussed earlier in the last class for editing the charts. 

For this, we'll need to assign plot to a variable so that we can use all the methods associated with the Axes object and Pyplot.

#### Changing chart parameters using Pyplot 

```python
sns.scatterplot(iris['sepal_length'], iris['petal_width'])
plt.title('Scatter Plot using Seaborn - Edit using Pyplot',size=20)                         # Using Pyplot functions to change Chart title
plt.xlabel('Sepal Length',size=15)                                      # Using Pyplot functions to change  x axis label
plt.ylabel('Petal Width',size=15)                                       # Using Pyplot functions to change  y axis label
```

#### Changing chart parameters using Axis Object 

```python
axs_obj = sns.scatterplot(iris['sepal_length'], iris['petal_width'])
axs_obj.set_title('Scatter Plot using Seaborn - Edit using Axes Object',size=20,color='blue')                         # Using Pyplot functions to change Chart title
axs_obj.set_xlabel('Sepal Length',size=15,color = 'maroon')                                      # Using Pyplot functions to change  x axis label
axs_obj.set_ylabel('Petal Width',size=15,color='maroon')                                       # Using Pyplot functions to change  y axis label
```

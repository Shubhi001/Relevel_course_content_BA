## Practice Question - 1

The data has been provided for ratings of different OTT shows. Make a bar graph as shown in the picture below.

```python
# Data

shows = ['Sacred Games', 'Leila', 'Ghoul', 'Mirzapur', 'The Family man', 'Paatal Lok']
ratings = [8.1,8,7.9,6,9,9.5]
```

#### Solution -

```python
plt.xlabel("Shows")                                    
plt.ylabel("Ratings")
plt.title("Ratings of OTT shows in 2020")
plt.xticks(rotation = 45)
plt.bar(shows, ratings, color='grey')
```

## Practice Question - 2

Additional to previous question, the making cost of shows has also been provide. 
THis time we need a combination of bar graph + line graph with their seprate y-axis. 
Please make a graph as shown in figure below. 

Hint - 
Use axes object to create a twin axes object using function twinx()


```python
# data

making_cost = [24,3,9,5,15,17,19]
```

#### Solution -

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()                                                                   # Creates a twin axes object
ax1.bar(shows, ratings, color='grey',label='ratings out of 10')
ax2.plot(np.arange(0,len(shows)+1), making_cost, c='black',label='Making cost')

ax1.set_xlabel('Shows')
ax1.set_title('Ratings and Making cost of famous OTTs',size = 14)
ax1.set_ylabel('Ratings out of 10',color='black',size=12 )
ax1.tick_params(axis='x', labelrotation = 45)

ax2.set_ylabel('Cost of making in Rs Crores',color='black',size=12)

plt.legend()
```

## Practice Question - 3

Fill between the area of 2 lines. The data for 2 line plots is provided. The desired output is as per sceenshot below. 

Hint - Use plt.fillbeween() function

```python
# data - 
x=np.linspace(0,4*np.pi,50)
line1 = np.sin(x)
line2 = 2*np.sin(x)
```

#### Solution -

```python
plt.plot(x,line1,'g--')
plt.plot(x,line2,'r--')
plt.fill_between(x,line1,line2,color='orange')
```

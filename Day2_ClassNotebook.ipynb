### 3. Bar chart (Plt.bar)

```python
days = ['Sunday','Momday','Tuesday','Wednesday','Thursday','Friday','Saturday']
sales = np.random.randint(50000,size=7)

plt.bar(days,sales,color = ['r','b','g','m','g','r','g'])
```

We can see that X-Axis tick labels are overlapping in above chart. This is a common problem when there are categorical
axes involved.

It can be resolved by rotating the labels as shown below

```python
days = ['Sunday','Momday','Tuesday','Wednesday','Thursday','Friday','Saturday']
sales = np.random.randint(50000,size=7)

plt.xticks(rotation = 90)                # Rotating Xticks by 90 degrees. Use -90 for anticlockwise


plt.bar(days,sales)


# The difference in values from above chart is because each time we run np.random.randint, the function generates random 
# numbers again
```

Another problem with this bar chart is lack of labels. As of now, we dont know what these numbers suggest and what

```python
days = ['Sunday','Momday','Tuesday','Wednesday','Thursday','Friday','Saturday']
sales = np.random.randint(50000,size=7)
plt.xticks(rotation = 90)      

# Adding more functions to define lables 
plt.xlabel('Days',color = 'blue')
plt.ylabel('Sales (In Rupees)', color = 'blue')
plt.title('Sales of Carpets in Week 13 of 2021',color='#a17d2a',size = '18')    # Hex color codes can be mentioned 

plt.bar(days,sales)
```

##### Variations in Bar charts

##### 3.1 Stacked Bar chart

Suppose there are sales of 2 shops in a given week.
If we are interested to see how total says were on each day, along with the split of each shop, a stacked bar chart
can be used as below- 

```python
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']

sales_shop1 = np.random.randint(50000,size=7)
sales_shop2 = np.random.randint(50000,size=7)
plt.xticks(rotation = 90)      


plt.bar(days,sales_shop1,label='Shop 1')
plt.bar(days,sales_shop2,label='Shop 2')
plt.legend()
```

##### 3.2 Dodged Bar Chart

But if we want to see the individual sales of shops, this will not be the desired output. 
We don't want the bar to be stacked.

This problem can be fixed as per below -

```python
days = ['Sunday','Momday','Tuesday','Wednesday','Thursday','Friday','Saturday']

sales_shop1 = np.random.randint(50000,size=7)       # generating random sales data for shop 1 for our example
sales_shop2 = np.random.randint(50000,size=7)       # generating random sales data for shop 1 for our example

x_indices = np.arange(0,len(days))                  # generating indices array for shop 1 bars 
x_indices_shop2 = np.arange(0,len(days))            # generating indices array for shop 2 bars 

width = 0.3                                         # defining a variable for bar width 

x_indices_shop2= x_indices_shop2+width               # Adjusted bars location for Shop 2 

print('For shop 1, bar will be plotted at positions -' + str(x_indices))         #Checking location of shop 1 bars
print('For shop 2, bar will be plotted at positions -' + str(x_indices_shop2))   #Checking location of shop 2 bars

plt.bar(x_indices,sales_shop1,width,label=1)          # Plottig shop 1 bars
plt.bar(x_indices_shop2,sales_shop2,width,label=2)    # Plottig shop 2 bars
plt.legend()

'''
Logic Explanation - 

By default, a second bar chart is plotted in same chart, the bars are plotted on same position, making it a stacked bar chart. 
Hence we needed to adjust the position of second bar series to start where the first bar ends. 
It would require following operations  
	Getting number of bars that will be plotted 
	Have an array from 0 to number of bars so that we have an index for our 1st series bar
	Have an another indices array for our 2nd series  but this time the indices will have to shift by the width of bars of first series so that the bars dont overlap. 
This is done by adding width of bars to first indices array
	We can plot the bar chart now. But there will be a small problem with x-ticks  By default the x-ticks will be centred on 1st bar series, we need to re-centre the xticks

'''
```

Now the only remaining problem is that xticks are not days names and xticks are not centred also.
Lets resolve these 2 issues by adjusting xticks and providing xlabels again.

Also, lets add all the axis labels, title etc to complete the chart. 

```python
days = ['Sunday','Momday','Tuesday','Wednesday','Thursday','Friday','Saturday']

sales_shop1 = np.random.randint(50000,size=7)
sales_shop2 = np.random.randint(50000,size=7)

x_indices = np.arange(0,len(days))
x_indices_shop2 = np.arange(0,len(days))

width = 0.3

x_indices_shop2= x_indices_shop2+width

print('For shop 1, bar will be plotted at positions -' + str(x_indices))
print('For shop 2, bar will be plotted at positions -' + str(x_indices_shop2))

plt.legend()

#--------------------------------Continuing below this from last cell------------------------------

#-------------------------------Adjusting xticks , providing xtick label as days
plt.xticks((x_indices+x_indices_shop2)/2, days)

#-------------------------------Providing chart and axis labels to complete the chart
plt.xlabel('Days',color = '#a17d2a')
plt.ylabel('Sales (In Rupees)', color = '#a17d2a')
plt.title('Sales of Carpets in Week 13 of 2021',color='#a17d2a',size = '18')  
plt.xticks(rotation = 90) 


plt.bar(x_indices,sales_shop1,width,label=1)
plt.bar(x_indices_shop2,sales_shop2,width,label=2)
```

### 7. Heatmap (plt.imshow & plt.his2D)

Heatmaps can be used to demonstrate data visually where 3 variables are used The 2 variables take up X and Y axis and third variable is represented by Intensity of color on the chart.

Heatmaps are applicable to very specific situations where there is a clear relationship between variables.
It is easy to make a really bad heatmap with no meaning, please be careful about that.

Heatmaps are definitely a bad option where categorical variables are involved.

Some good examples of situations where heatmaps can be used/have been used are - 

* Temperature(colored) across longitude and latitude
* Probability of a missing plane(colored) across longitude and latitude
* Correlation(Colored) among different quantitative variables present in a data


Lets make a sample heatmap below for a random data. We'll have 2 sets of 1000 random numbers all in the range between 0 to 1 (total 2000 numbers).

The color of a section in Heatmap will depict the frequency of numbers falling in that bracket while the range of 2 sets are represented on X and Y axis.


```python
# Creatind random data for plotting a heatmap
data = pd.DataFrame(np.random.rand(1000,2),columns=list('AB'))

_ = plt.hist2d(data['A'],data['B'], bins=10)

# plt.colorbar adds a legend 
plt.colorbar()
plt.title('Heatmap showing frequencies of values by Color')

```

### 8. Subplots ( plt.subplot )

Subplots, as the name suggests, are multiple plots on the same canvas.

These are extremely useful to show multiple charts in same frame, especially if those are interlinked and a story can be told by the looking at a picture collectively.


Subplot function takes in 3 parameters -

*	1st one denotes number of rows
*	2nd denotes number of columns
*	3rd one selects the particular subplot

For example, subplot (2,2,3) means that we are defining 2 rows and 2 columns of subplots, which means there will be total 2*2 = 4 subplots on the canvas. 3 in last parameter means that we are selecting the 3rd subplot, and all the functions applied after this will work on 3rd subplot.

We can toggle between subplots by mentioning the correct subplot code like (2,2,1) or (2,2,2) or (2,2,3) or (2,2,4)
In this way we can make different charts on same canvas.




```python
plt.subplot(2,2,1)                                     # Selecting subplot 1 (Top Left)
plt.plot(x_int,y_int,'r--')                
plt.subplot(2,2,2)                                      # Selecting subplot 2 (Top Right)                                    
plt.plot(x_int,y_int,'g*--')
plt.subplot(2,2,3)                                     # Selecting subplot 3 (Bottom Left)
plt.plot(x_triangles,y_triangles,'bo')
plt.subplot(2,2,4)                                     # Selecting subplot 4 (Bottom Right)
plt.scatter(x_triangles,y_triangles,marker='v')


# You might notice use of r--, g-- etc above
# these are alternate ways to define colors and line patterns in plot function 
# you can always do a shift+tab by placing cursor on any function and learn about these methods by scrolling down 
# For example, these r-- and g-- are mentioned in Markers section for plot function help window
```

By the way, by default Pyplot assumes a canvas as (1,1,1) subplot. Which means a single plot

### 9. Axes Object

We had learnt about OOPS yesterday. You might wonder that we have not used any of that today, so what is the significance and why did we go through those concepts.

In reality, all the functions that we have used today are using OOPS at the backend. Its just that the Pyplot scripting layer is so great at a high level that we do not even get a hint of how much work is going on at the backend.
In background, pyplot is maintaining track of multiple object a figure contains, and modifying different components as per our execution.

For example, a simple task of assigning the X-axis label involves checking if a figure exists or not (and creating a new one in case it doesn't), finding the right object which deals with axes, then finding the child object which keeps track of X-axis, then subsequently changing the value of label. In the end, it also ensures multiple different objects are modified accordingly (like size of label, font etc.) so that the figure is rendered properly without any error.


One of the key objects Pyplot interacts a lot with is Axes object. Lets dive a bit into it and see what all different child object comes under axes object.

```python
# get the current axes
ax_object = plt.gca()

# Let's see what all child objects this axes object contains
ax_object.get_children()
```

As we can see, it contains about 10 child components- 
1. 4 spine line objects which keep track of area boundaries
2. 1 X Axis object
3. 1 Y Axis object
4. 3 Text objects
5. 1 rectangle canvas object

Although we have already learnt to use pyplot for most of our needs, but we can also perform all those functions
directly with the help of Axes object too without the use of Pyplot. 

Again, we are discussing this because you'll see a lot of discussion about these on forums like stackoverflow 
and its always good to be aware of all the different approaches. 

```python
plt.plot(x_int,y_int, marker='s', label = 'data 1')
plt.scatter(x_triangles,y_triangles, c= 'maroon', label = 'data 2')

axes_object = plt.gca()
axes_object.set_title('Chart Title set by Axes Object',
                         fontweight ="bold")
axes_object.set_xlabel('X Label set by Axes Object')
axes_object.set_ylabel('Y Label set by Axes Object')
axes_object.legend()
```

#### Axes Object in case of Subplots

In case of subplots, we can get separate axes objects for each subplot using syntax given below.
Each subplot then can be individually modified using the related axes object.

For 99% of the time, we will not need to use this. But knowing that such objects exists below Pyplot helps a lot in understanding and mastering Pyplot.


```python
x_triangles = np.arange(1,21)
y_triangles = np.random.randint(1000,size=20)        
x_int = np.arange(1,21)                 # creates an array of 20 elements of  integers from 1 to 20
y_int = np.random.randint(1000,size=20) # creates an array of 20 elements of random integers between 0-1000

#-------Getting Axes Objects for each Subplot and assigning each to different variable for subsequent usage---------------

fig, ax = plt.subplots(2, 2, sharex=True)       # Assign figure object to fig and Axes objects to ax
ax1 = ax[0,0]                                   # Axes object for subplot 1 
ax2 = ax[0,1]                                   # Axes object for subplot 2
ax3 = ax[1,0]                                   # Axes object for subplot 3 
ax4= ax[1,1]                                    # Axes object for subplot 4 




#--------------------------Plotting data in Plot 1 and modifying its labels---------------------------
ax1.plot(x_int,y_int,'r--')                 
ax1.set_title('Chart Title set by Axes Object')
ax1.set_xlabel('X Label set by Axes Object')
ax1.set_ylabel('Y Label set by Axes Object')

#--------------------------Plotting data in Plot 2 and modifying its labels---------------------------
ax2.plot(x_int,y_int,'g*--')
ax2.set_title('Chart Title set by Axes Object')
ax2.set_xlabel('X Label set by Axes Object')
ax2.set_ylabel('Y Label set by Axes Object')
    
#--------------------------Plotting data in Plot 3 and modifying its labels---------------------------
ax3.plot(x_triangles,y_triangles)
ax3.set_title('Chart Title set by Axes Object')
ax3.set_xlabel('X Label set by Axes Object')
ax3.set_ylabel('Y Label set by Axes Object')
               
#--------------------------Plotting data in Plot 4 and modifying its labels---------------------------

ax4.set_xlabel('X Label set by Axes Object')
ax4.set_ylabel('Y Label set by Axes Object')
ax4.set_title('Chart Title set by Axes Object')
ax4.scatter(x_triangles,y_triangles,marker='v')
```

We can see that we have been able to modify the parameters using axes object.
But it is clearly visible that there is a lot of overlapping going on.
